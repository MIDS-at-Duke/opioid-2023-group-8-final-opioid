```python
csv_chunk = pd.read_table(
    "../Data/arcos_all_washpost.tsv", chunksize=50_000, low_memory=False
)
```

```python
Washington = []
for i, chunk in enumerate(csv_chunk):
    append_chunk = chunk.loc[chunk["BUYER_STATE"] == "WA"]
    Washington.append(append_chunk)
data_selected = pd.concat(Washington)
```

```python
data_selected["NDC_NO"] = data_selected["NDC_NO"].astype(str)
```

```python
data_selected.to_parquet("../Data/Washington.parquet")
```
